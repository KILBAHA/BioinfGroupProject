```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import all TARGET data z-score files from cBioPortal
mRNA_Zscores = pd.read_csv('../nbl_target_2018_pub/data_mRNA_median_Zscores.txt',  sep='\t')
mRNA_all_sample_Zscores = pd.read_csv('../nbl_target_2018_pub/data_mRNA_median_all_sample_Zscores.txt',  sep='\t')
RNA_Seq_mRNA_Zscores = pd.read_csv('../nbl_target_2018_pub/data_RNA_Seq_mRNA_median_Zscores.txt',  sep='\t')
RNA_Seq_mRNA_all_sample_Zscores = pd.read_csv('../nbl_target_2018_pub/data_RNA_Seq_mRNA_median_all_sample_Zscores.txt',  sep='\t')
```

```python
#Plot histograms of z-scores for each file
fig, axs = plt.subplots(2, 2)
axs[0,0].hist(mRNA_Zscores.iloc[:,2:].values.astype(float).flatten(),bins=100)
axs[0,0].set_title('mRNA_Zscores: ' + str(mRNA_Zscores.shape), fontsize=10)
axs[0,1].hist(mRNA_all_sample_Zscores.iloc[:,2:].values.astype(float).flatten(),bins=100,color='orange')
axs[0,1].set_title('mRNA_all_sample_Zscores: ' + str(mRNA_all_sample_Zscores.shape), fontsize=10)
axs[1,0].hist(RNA_Seq_mRNA_Zscores.iloc[:,2:].values.astype(float).flatten(),bins=100,color='green')
axs[1,0].set_title('RNA_Seq_Zscores: ' + str(RNA_Seq_mRNA_Zscores.shape), fontsize=10)
axs[1,1].hist(RNA_Seq_mRNA_all_sample_Zscores.iloc[:,2:].values.astype(float).flatten(),bins=100,color='red')
axs[1,1].set_title('RNA_Seq_all_sample_Zscores: ' + str(RNA_Seq_mRNA_all_sample_Zscores.shape), fontsize=10)
fig.tight_layout()
plt.show()
```

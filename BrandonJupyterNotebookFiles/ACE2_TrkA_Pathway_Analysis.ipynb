# Compare prevalence of activated TrkA pathway in samples stratified by ACE2 expression

```python
#using z-score cutoffs of -0.5 and 0.5 for activated pathway
ACE2_groups = stratifyDataOnGeneExpression2('ACE2',-0.5,0.5)

groups_count = [0,0,0]
i=0
for group in ACE2_groups:
    for sample in group:
        tp53_score = z_scores.loc[z_scores['Hugo_Symbol']=='TP53'][sample].values[0]
        ptpn6_score = z_scores.loc[z_scores['Hugo_Symbol']=='PTPN6'][sample].values[0]
        ntrk1_score = z_scores.loc[z_scores['Hugo_Symbol']=='NTRK1'][sample].values[0]
        if(tp53_score < -0.5 and ptpn6_score < -0.5 and ntrk1_score > 0.5):
            groups_count[i]+=1
    i+=1

low_per = groups_count[0]/len(ACE2_groups[0]) * 100
intermediate_per = groups_count[1]/len(ACE2_groups[1]) * 100
high_per = groups_count[2]/len(ACE2_groups[2]) * 100
print("Number of Low ACE2 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[0]) + "/" + str(len(ACE2_groups[0])) + " (" + str(low_per) + "%)")
print("Number of Intermediate ACE2 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[1]) + "/" + str(len(ACE2_groups[1])) + " (" + str(intermediate_per) + "%)")
print("Number of High ACE2 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[2]) + "/" + str(len(ACE2_groups[2])) + " (" + str(high_per) + "%)")



```

# Analyze role of ACE2 on EFS in samples with TrkA pathway activation

```python
#using z-score cutoff of 0 for activated pathway
activated_samples = []

for sample in z_scores.iloc[:,2:].columns:
    tp53_score = z_scores.loc[z_scores['Hugo_Symbol']=='TP53'][sample].values[0]
    ptpn6_score = z_scores.loc[z_scores['Hugo_Symbol']=='PTPN6'][sample].values[0]
    ntrk1_score = z_scores.loc[z_scores['Hugo_Symbol']=='NTRK1'][sample].values[0]
    if(tp53_score < 0 and ptpn6_score < 0 and ntrk1_score > 0):
        activated_samples.append(sample)

nonactivated_samples = list(set(z_scores.iloc[:,2:].columns) - set(activated_samples))
```

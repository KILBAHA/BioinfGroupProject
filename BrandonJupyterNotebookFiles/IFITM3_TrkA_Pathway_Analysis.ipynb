# Compare prevalence of activated TrkA pathway in samples stratified by IFITM3 expression

```python
#using z-score cutoff of 0 for activated pathway
#calculate prevalence in low, intermediate, and high groups
IFITM3_groups = stratifyDataOnGeneExpression2('IFITM3',-0.5,0.5)

groups_count = [0,0,0]
i=0
for group in IFITM3_groups:
    for sample in group:
        tp53_score = z_scores.loc[z_scores['Hugo_Symbol']=='TP53'][sample].values[0]
        ptpn6_score = z_scores.loc[z_scores['Hugo_Symbol']=='PTPN6'][sample].values[0]
        ntrk1_score = z_scores.loc[z_scores['Hugo_Symbol']=='NTRK1'][sample].values[0]
        if(tp53_score < 0 and ptpn6_score < 0 and ntrk1_score > 0):
            groups_count[i]+=1
    i+=1

low_per = groups_count[0]/len(IFITM3_groups[0]) * 100
intermediate_per = groups_count[1]/len(IFITM3_groups[1]) * 100
high_per = groups_count[2]/len(IFITM3_groups[2]) * 100
print("Number of Low IFITM3 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[0]) + "/" + str(len(IFITM3_groups[0])) + " (" + str(low_per) + "%)")
print("Number of Intermediate IFITM3 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[1]) + "/" + str(len(IFITM3_groups[1])) + " (" + str(intermediate_per) + "%)")
print("Number of High IFITM3 samples with Low TP53&PTPN6/High NTRK1: " + str(groups_count[2]) + "/" + str(len(IFITM3_groups[2])) + " (" + str(high_per) + "%)")



```

```python
#Chi-Squared Test comparing all three groups
contingency_table = [[73-16,100-27,76-12],[16,27,12]]
stat, p, dof, expected = chi2_contingency(contingency_table)
print("p-value from Chi-Squared Test: " + str(p))
```

# Analyze role of IFITM3 on EFS in samples with TrkA pathway activation

```python
#using z-score cutoff of 0 for activated pathway
activated_samples = []

for sample in z_scores.iloc[:,2:].columns:
    tp53_score = z_scores.loc[z_scores['Hugo_Symbol']=='TP53'][sample].values[0]
    ptpn6_score = z_scores.loc[z_scores['Hugo_Symbol']=='PTPN6'][sample].values[0]
    ntrk1_score = z_scores.loc[z_scores['Hugo_Symbol']=='NTRK1'][sample].values[0]
    if(tp53_score < 0 and ptpn6_score < 0 and ntrk1_score > 0):
        activated_samples.append(sample)

nonactivated_samples = list(set(z_scores.iloc[:,2:].columns) - set(activated_samples))
```

```python
low_IFITM3 = []
intermediate_IFITM3 = []
high_IFITM3 = []
for sample in activated_samples:
    IFITM3_score = z_scores.loc[z_scores['Hugo_Symbol']=='IFITM3'][sample].values[0]
    if(IFITM3_score < -0.5):
        low_IFITM3.append(sample)
    elif(IFITM3_score > -0.5 and IFITM3_score < 0.5):
        intermediate_IFITM3.append(sample)
    else:
        high_IFITM3.append(sample)
```

```python
data = [low_IFITM3,intermediate_IFITM3,high_IFITM3]
compareEFS2(data,'EFS in samples with active TrkA pathway stratified by IFITM3 expression', ['Low','Intermediate','High'], 't-test_ind')


```

```python
kaplanmeierEFS2(data, 'EFS in samples with active TrkA pathway stratified by IFITM3 expression', ['Low','Intermediate','High'])



```

```python
not_high_IFITM3 = list(set(activated_samples) - set(high_IFITM3))
data=[not_high_IFITM3,high_IFITM3]
compareEFS2(data,'EFS in samples with active TrkA pathway stratified by IFITM3 expression', ['Remaining Samples','High IFITM3 samples'], 't-test_ind')


```

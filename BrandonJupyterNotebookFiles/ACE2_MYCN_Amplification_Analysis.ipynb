# Compare ACE2 Expression and MYCN Amplification

```python
labels = ['low','intermediate','high']
for i in range(len(ACE2_groups)):
    samples = clinical_sample_trimmed.loc[clinical_sample_trimmed['Sample Identifier'].isin(ACE2_groups[i])]
    amplified_MYCN_count = sum(samples['MYCN']=='Amplified')
    amplified_MYCN_prop = amplified_MYCN_count/len(samples)
    print("Number of MYCN amplified tumors in " + labels[i] + " ACE2 expression group: " + str(amplified_MYCN_count) + "/" + str(len(samples)))
    print("Proportion of MYCN amplified tumors in " + labels[i] + " ACE2 expression group: " + str(amplified_MYCN_prop))
    
    
    
```

```python
#Chi-Squared Test comparing all three groups
contingency_table = [[63,72,46],[16,25,27]]
stat, p, dof, expected = chi2_contingency(contingency_table)
print("p-value from Chi-Squared Test: " + str(p))
```

```python
#Chi-Squared Test comparing high ACE2 expressing samples with rest
contingency_table = [[63+72,46],[16+25,27]]
stat, p, dof, expected = chi2_contingency(contingency_table)
print("p-value from Chi-Squared Test: " + str(p))
```

# Comparing EFS based on both ACE2 expression and MYCN amplification status

### Comparing EFS in MYCN nonamplified samples based on ACE2 expression

```python
#Stratify samples based on MYCN amplification status
MYCN_groups = stratifyCategoricalSampleData('MYCN', 'Not Amplified', 'Amplified')
```

```python
#Find z-scores for MYCN nonamplified samples
nonamplified_MYCN_columns = ['Hugo_Symbol','Entrez_Gene_Id'] + list(set(MYCN_groups[0]).intersection(z_scores.columns))
nonamplified_MYCN_z_scores = z_scores[nonamplified_MYCN_columns]

```

```python
#Obtain the ACE2 z-scores in nonamplified MYCN samples
ACE2_MYCN_nonamplified_z_scores = nonamplified_MYCN_z_scores.loc[nonamplified_MYCN_z_scores['Hugo_Symbol']=='ACE2'].iloc[:,2:].transpose()
#Find samples for which the z-scores are < -0.5
low_ACE2_MYCN_nonamplified_samples = ACE2_MYCN_nonamplified_z_scores.loc[ACE2_MYCN_nonamplified_z_scores.iloc[:,0] < -0.5].index
#Find samples for which z-scores are between -0.5 and 0.5
intermediate_ACE2_MYCN_nonamplified_samples = ACE2_MYCN_nonamplified_z_scores.loc[ACE2_MYCN_nonamplified_z_scores.iloc[:,0] > -0.5]
intermediate_ACE2_MYCN_nonamplified_samples = intermediate_ACE2_MYCN_nonamplified_samples.loc[intermediate_ACE2_MYCN_nonamplified_samples.iloc[:,0] < 0.5].index
#Find samples for which the z-scores are > 0.5
high_ACE2_MYCN_nonamplified_samples = ACE2_MYCN_nonamplified_z_scores.loc[ACE2_MYCN_nonamplified_z_scores.iloc[:,0] > 0.5].index

```

```python
data = [low_ACE2_MYCN_nonamplified_samples, intermediate_ACE2_MYCN_nonamplified_samples, high_ACE2_MYCN_nonamplified_samples]
compareEFS2(data,'EFS in MYCN nonamplified samples based on ACE2 expression',['Low','Intermediate','High'],'t-test_ind')
       
       
```

```python
kaplanmeierEFS2(data, 'EFS in MYCN nonamplified samples stratified by ACE2 expression', ['Low','Intermediate','High'])


```

```python
data = [low_ACE2_MYCN_nonamplified_samples, high_ACE2_MYCN_nonamplified_samples]
compareEFS2(data,'EFS in MYCN nonamplified samples based on ACE2 expression',['Low','High'],'t-test_ind')
kaplanmeierEFS(data,'Low','High')

       
```

```python
not_low_ACE2_MYCN_nonamplified_samples = ACE2_MYCN_nonamplified_z_scores.loc[ACE2_MYCN_nonamplified_z_scores.iloc[:,0] > -0.5].index
data = [low_ACE2_MYCN_nonamplified_samples, not_low_ACE2_MYCN_nonamplified_samples]
compareEFS(data, 'EFS stratified by ACE2 expression', 'Low ACE2 Patients', 'Remaining Patients')

```

# TP53/PTPN6/NTRK1 Pathway Analysis

```python
#using z-score cutoff of 0 for activated pathway
activated_samples = []

for sample in z_scores.iloc[:,2:].columns:
    tp53_score = z_scores.loc[z_scores['Hugo_Symbol']=='TP53'][sample].values[0]
    ptpn6_score = z_scores.loc[z_scores['Hugo_Symbol']=='PTPN6'][sample].values[0]
    ntrk1_score = z_scores.loc[z_scores['Hugo_Symbol']=='NTRK1'][sample].values[0]
    if(tp53_score < 0 and ptpn6_score < 0 and ntrk1_score > 0):
        activated_samples.append(sample)

nonactivated_samples = list(set(z_scores.iloc[:,2:].columns) - set(activated_samples))

data = [activated_samples,nonactivated_samples]
```

```python
compareExpression2('BSG',data,'TrkA Pathway Activation',['Activated',"Not activated"],'t-test_ind')
#TP53
geneScatter('BSG','TP53')
#PTPN6
geneScatter('BSG','PTPN6')
#NTRK1
geneScatter('BSG','NTRK1')
```

```python
compareExpression3('BSG',data,'TrkA Pathway Activation','CD147 Expression stratified by TrkA Pathway Activation',['Activated',"Not activated"],'t-test_ind')


```

```python
compareExpression2('DPP4',data,'TrkA Pathway Activation',['Activated',"Not activated"],'t-test_ind')
#TP53
geneScatter('DPP4','TP53')
#PTPN6
geneScatter('DPP4','PTPN6')
#NTRK1
geneScatter('DPP4','NTRK1')
```

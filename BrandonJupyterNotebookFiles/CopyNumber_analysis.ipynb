```r
library(copynumber)
library(tibble)
```

```r
cna_data = read.table('../nbl_target_2018_pub/data_cna_hg19_copy.txt',  sep='\t', header = TRUE, stringsAsFactors = FALSE)
cna_data['chrom'] = replace(cna_data['chrom'], cna_data['chrom'] == 'X', '23')
cna_data['chrom'] = as.integer(cna_data$chrom)
colnames(cna_data) = c('sampleID','chrom','start.pos','end.pos','n.probes','mean')
cna_data = add_column(cna_data, arm = 0, .after = "chrom")
```

```r
plotAberration(segments=cna_data,thres.gain=0.2,chrom=c(12))
```

```r
plotChrom(segments = cna_data,sample=c(42),chrom=17)
```
